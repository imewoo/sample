# where

```python
def categorizing_by_5000(data):
    data=data.where(lambda x: x>=5000, 0)
    data=data.where(lambda x: x<5000, 1)
    return data
```

# apply

```python
ktx_df.iloc[:, :-2].apply(categorizing_by_5000) #iloc (전체를 가져오고, column은 뒤꺼 하나 뺴고)
```
