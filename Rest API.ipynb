```python
url='http://apis.data.go.kr/B500001/drghtIdexSpiAnals/analsInfoList'
params={
    'hjdCd':'1168058000',
    'stDt':'20190501',
    'edDt':'20190531',
    'numOfRows':10,
    'pageNo':1,
    'serviceKey':'jd6skVkLMALtJIyfMndF62XOQlDZ+iI4Bbji/raDq+7/1kU9mcmsPDrOMLqoZzVPQA0eaakjXKw24dCG6mPGeQ=='
}
```

```python
resp=requests.get(url,params=params)
resp.text
```

```python
soup=BeautifulSoup(resp.text, 'lxml')
soup
```

```python
header=soup.find('header')
body=soup.find('body')
```

```python
result_code=header.find('resultcode').text
result_msg=header.find('resultmsg').text
result_code, result_msg
```

```python
num_of_rows=body.find('numofrows').text
page_no=body.find('pageno').text
total_count=body.find('totalcount').text
num_of_rows, page_no, total_count
```

```python
items=body.find_all('item')

result_list=[]
for item in items:
    result_list.append({
        'anldt':item.find('anldt').text,
        'anlrst':item.find('anlrst').test,
        'anlval':item.find('anlval').text,
        'dv':item.find('dv').text,
        'hjdcd':item.find('hjdcd').text
    })
result_list
```

```python
import datetime
now=datetime.datetime.now()
today=now.strftime('%Y%n%d')
```

```python
url='http://apis.data.go.kr/B500001/drghtIdexSpiAnals/analsInfoList'
params={
    'hjdCd':'1168058000',
    'stDt':'20100501',
    'edDt':today,
    'numOfRows':1000000,
    'pageNo':1,
    'serviceKey':'jd6skVkLMALtJIyfMndF62XOQlDZ+iI4Bbji/raDq+7/1kU9mcmsPDrOMLqoZzVPQA0eaakjXKw24dCG6mPGeQ==',
    '_type':'json'
}

resp=requests.get(url,params=params)
resp
```

```python
data=resp.json()
data
```
